```python
import pandas as pd
import collections
```

```python
df = pd.read_csv('data_csv.csv')
```

```python
df
```

```python
df['Name'] = df['Name'].astype(str)
```

```python
df['Code'] = df['Code'].astype(str)
```

```python
df
```

### Converting into Dictionary

```python
dictt = df.set_index('Code')['Name'].to_dict()
```

### Ordered Dict AT keys

```python
od = collections.OrderedDict(sorted(dictt.items()))
```

```python
od
```

```python
Country_code_1 = input("Enter your value 1: ")
```

```python
Country_code_2 = input("Enter your value 2: ")
```

### Countries between IN and US

```python
countries = []
for key in od.keys():
    if(key > Country_code_1) & (key < Country_code_2):
        countries.append(od[key])
    
```

```python
countries.sort()
```
