```python
df = pd.read_csv('company_sales_data.csv', index_col = 'month_number')
```

### a) Show the Total profit of all months using a line plot

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Total profit vs month")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(x = df.index,y=df['total_profit'])



# Add label for horizontal axis
plt.xlabel("Month")
plt.ylabel("Profit")
```

### b) Show all product sales data using a multiline plot

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Different items sold over time")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=df.iloc[:,1:-2])



# Add label for horizontal axis
plt.xlabel("Month")
plt.ylabel("No. of items")
```

### c) Calculate total sales data for last year for each product and show it using a Pie chart

```python
sum = df.sum(axis = 0, skipna = True)
```

```python
dff= pd.DataFrame(sum)
```

```python
dff = dff.iloc[:-2,]
```

```python
# Creating plot 
fig = plt.figure(figsize =(10, 7)) 
plt.pie(dff[0], labels = dff.index) 
  
# show plot 
plt.show() 
```

```python
data = dff[0].tolist()
```

```python
data
```

```python
products = dff.index
```

```python
products = products.tolist()
```

```python
products
```

```python

  

  
  
# Creating explode data 
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0) 
  
# Creating color parameters 
colors = ( "orange", "cyan", "brown", 
          "grey", "indigo", "beige") 
  
# Wedge properties 
wp = { 'linewidth' : 1, 'edgecolor' : "green" } 
  
# Creating autocpt arguments 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n({:d} )".format(pct, absolute) 
  
# Creating plot 
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data), 
                                  explode = explode,  
                                  labels = products, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="magenta")) 
  
# Adding legend 
ax.legend(wedges, products, 
          title ="products", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Customizing pie chart") 
  
# show plot 
plt.show() 
```
